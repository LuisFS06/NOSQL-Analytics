```python
#Segunda Parte: Acceder a la pagina de interés:

#En este caso se decidio utilizar Nike
link = "https://www.instagram.com/jet.costarica/"

driver.get(link)
res = requests.get(link,headers ={'User-Agent': 'Mozilla/5.0'})

soup = BeautifulSoup(res.text, "html.parser")
tag = soup.find("meta", property="og:description")
tag = tag.get("content", None)

#Agarrar todos los links de la pagina web 
links = [] #lista en el cual se guardaran todos los links de cada post

number_of_posts = int(tag.split(',')[-1].split('-')[0].split()[0]) #Cantidad de post en la página

while True:
    hrefElements = WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.XPATH, "//div[@class='_aabd _aa8k  _al3l']/a")))

    elements_link = [x.get_attribute("href") for x in hrefElements]
    for link in elements_link:
        if link not in links:
            links.append(link)

    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    driver.implicitly_wait(5)

    if len(links) >= number_of_posts:
          break
            
links
```
